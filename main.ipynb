1. Получить двумерные спектры изображений
2. Проанализировать спектры изображений с разной структурой

```python
from skimage import color, data, io
from matplotlib import pyplot as plt

import numpy as np
```

```python
def print_pic(pic):
  io.imshow(pic)
  plt.show()
```

```python
# выгрузка изображений из библиотеки, для дальнейшей работы с ними
pic_list = [color.rgb2gray(data.coffee()),
            color.rgb2gray(data.text()),
            color.rgb2gray(data.page()),
            color.rgb2gray(data.rocket()),
            color.rgb2gray(data.checkerboard())]

[print_pic(x) for x in pic_list]
```

```python
def create_two_dimensional_fourier_spectrum(pic):
  print("==================================")
  print_pic(pic)

  fourier_spectrum = np.log(1+np.abs(np.fft.fftshift(np.fft.fft2(pic))))
  return fourier_spectrum

print_pic(create_two_dimensional_fourier_spectrum(color.rgb2gray(data.checkerboard()))) 
```

```python
# анализ
[print_pic(create_two_dimensional_fourier_spectrum(x))  for x in pic_list]
```
